```python
def assert_consistent_sense_data(now_sense_data, pre_sense_data):
    assert_sense(now_sense_data['sense'], pre_sense_data['sense'])
    assert_extra_sents(now_sense_data['extra_sents'], pre_sense_data['extra_sents'])
    assert now_sense_data['guideword'] == pre_sense_data['guideword']
    return

def assert_sense(now_sense_list, pre_sense_list):
    assert len(now_sense_list) == len(pre_sense_list)
    for now_sense, pre_sense in zip(now_sense_list, pre_sense_list):
        assert now_sense['en_def'] == pre_sense['en_def']
        assert now_sense['ch_def'] == pre_sense['ch_def']
        assert ((now_sense['level'] == pre_sense['level'])
                or (now_sense['level'] == '' and pre_sense['level'] == 'none'))
        now_examples = []
        for ne in now_sense['examples']:
            if ne['en']:
                now_examples.append(ne['en'])
            if ne['ch']:
                now_examples.append(ne['ch'])
        assert len(now_examples) == len(pre_sense['examples'])
        for ne, pe in zip(now_examples, pre_sense['examples']):
            assert ne == pe
    return

def assert_extra_sents(now_extra_sents, pre_extra_sents):
    assert len(now_extra_sents) == len(pre_extra_sents)
    for ns, ps in zip(now_extra_sents, pre_extra_sents):
        assert ns == ps
    return
```

```python
assert_consistent_sense_data(now_data[check_vocabs[0]]['noun'][1],
                             pre_data[check_vocabs[0]]['noun'][1])
```

```python
skip_vocabs = ['miss']
for vocab in check_vocabs:
    if vocab in skip_vocabs:
        continue
    assert len(now_data[vocab]) >= len(pre_data[vocab])
    for pos in pre_data[vocab]:
        for now_sense_data, pre_sense_data in zip(now_data[vocab][pos],
                                                  pre_data[vocab][pos]):
            if now_sense_data['phrase']:
                continue
            assert_consistent_sense_data(now_sense_data, pre_sense_data)
```
